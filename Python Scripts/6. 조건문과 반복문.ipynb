- if 문의 조건이 맞는지 확인
- if 문의 조건이 참인 경우, if의 body를 실행하고 if...elif...else statement를 종료
    - 이 때, elif와 else의 조건이 맞는지 여부는 확인하지 않음

- if 문의 조건이 거짓인 경우, elif의 조건의 참/거짓 여부를 확인
- elif의 조건이 참이라면, elif의 body를 실행하고 if...elif...else statement를 종료
    - 이 때, 첫 번째 elif를 제외한 elif와 else의 조건이 맞는지 여부는 확인하지 않음
    
$$\vdots$$

- if와 elif의 모든 조건이 거짓일 경우, else의 body를 실행함


```python
x = 20

if x == 12:
    print('x is 12.')
elif x == 14:
    print('x is 14.')
elif x == 16:
    print('x is 16.')
else:
    print('x is neither 12 nor 14 nor 16.')
```

```python
x = 20
if x == 20:
    print('if: x is 20.')
elif x == 20:
    print('elif: x is 20.')
else:
    print('x is not 20.')

print('Body of elif condition is not executed.')
```

# multiple conditions
- and 와 or을 이용해 조건식들을 조합할 수 있다
- and는 &와 다르고, or은 |와 다르다
    - and, or (Boolean / logical operator)
    - &, | (bitwise operator)
    
- 두 변수 값이 같은지 다른지를 확인할 때는 "=="과 "!="을 사용한다
    - "is"와 "is not"은 "&, |"에 대응하는 개념이다
    

- 일반적으로 원하는 조건을 얻기 위해서는 and, or, ==, !=를 사용

```python
x = 3
print(x == 3 and x < 10)
print(x == 3 & x < 10)
print('-'*80)

print(x < 10 and x < 20) # desired
print(x < 10 & x < 20) # undesired


print('-'*80)
print(x == 3)
print(x != 3)
```

# for loop
- sequence (list, tuple, dict, set, str)에 대해 iterate을 할 때 사용

- break, continue, pass 등의 keyword를 사용 가능

## for loop on list, tuple, dict, set, str

```python
mylist = [1,2,3]
for x in mylist:
    print(x)

print('-'*80)
mytuple = (1,2,3)
for x in mytuple:
    print(x)
    
print('-'*80)    
mydict = {'a':1, 'b':2, 'c':3}
for key in mydict: # iterate through key
    print(key, mydict[key])

print('-'*80)    
for item in mydict.items(): # iterate through item
    print(item)
    
print('-'*80)    
myset = set([1,2,3])
for x in myset:
    print(x)

print('-'*80)    
mystr = 'apple'
for x in mystr:
    print(x)    
```

## Tuple assignment in for loops

```python
mytuple = [(1,2), (3,4)]
for (a,b) in mytuple:
    print(f'a:{a}, b:{b}')
 
    
print('-'*80)    
mytuple = [([1,2], 3), ((4, 5), 6)]
for ((a, b), c) in mytuple:
    print(a, b, c, sep=',')

print('-'*80)        
for (a, b) in mytuple:
    print(f'a: {a}, b: {b}')
```

## Nested for loops

```python
# for given outer for loop, inner for loops are executed
for x in range(2):
    for y in range(3):
        print('x is:', x)
        print('y is:', y)
        print('')
```

## loop multiple sequence in parallel: zip
- zip function을 사용해 multiple sequence에 대한 parallel for loop가 가능

```python
mylist1 = range(8, 11, 1)
mylist2 = range(3, 6, 1)
mylist3 = ['a', 'b', 'c']
print(zip(mylist1, mylist2, mylist3)) # zip() returns an object
print(list(zip(mylist1, mylist2, mylist3))) # list() to see the result
```

```python
for (x, y, z) in zip(mylist1, mylist2, mylist3): # tuple assignment to unpack the zip result
    print(f'x:{x}, y:{y}, z:{z}', sep=',')
```

# while loop
- while condition 이 참인 동안 while loop의 body를 실행한다
- while condition은 body를 실행한 후에 참/거짓 여부를 다시 확인한다

```python
from IPython.display import Image
display(Image("http://www.trytoprogram.com/images/python_while_loop.jpg", width=400))
```

```python
x = 0
while x < 5:
    x = x + 1
    print(f'x: {x}')
else:
    print('run if break is not hit')
```

```python
x = 0
while x < 5:
    x = x + 1
    if x==2 or x==3:
        continue

    print(f'x: {x}')
        
    if x==5:
        break
else:
    print('run if break is not hit')
```

Example) 100 이하의 자연수 중 2의 배수도 아니고, 3의 배수도 아닌 수들의 합은?

```python
# Answer 1
s = 0
for x in range(101):
    if x % 2 != 0 and x % 3 != 0:
        s = s + x
        
print(s)
```

```python
# Answer 2
l = []
for x in range(101):
    if x % 2 != 0 and x % 3 != 0:
        l.append(x)
        
print(l)
print(sum(l))
```

Q) 100이하의 자연수 중 4의 배수이거나 5의 배수인 수들의 합은?

```python
# Answer
l = []
for x in range(101):
    if x % 4 == 0 or x % 5 == 0:
        l.append(x)
        
print(l)
print(sum(l))
```

## List comprehension
- for loop과 비슷한 기능 제공
- for loop보다 가독성이 높은 문법을 제공
- manual for loop보다 더 빨리 실행될 수 있음

```python
mylist = [1, 2, 3, 4, 5]
for i in range(5):
    mylist[i] += 10

print(mylist)

mylist = [1, 2, 3, 4, 5]
mylist_ = [x + 10 for x in mylist] # equivalent to above
print(mylist_)
```

### What listcomp does internally

```python
mylist = list(range(1, 6)) # [1, 2, 3, 4, 5]
mylist__ = []
for x in mylist:
    mylist__.append(x+10)
    
print(mylist__)
```

## More on listcomp
### if with listcomp

```python
a = range(1, 13)
b = [x for x in a if x % 4 == 0] # if x is a multiple of 4
print(b)
```

### nested loops in listcomp

```python
mylist = []
for x in 'ab':
    for y in 'xy':
        mylist.append(x+y)
        
print(f'nested for: {mylist}')

mylist = [x+y for x in 'ab' for y in 'xy']
print(f'listcomp: {mylist}')
```

# try-except statement
- try-except를 통해 error를 handle할 수 있음
- 예상치 못한 error가 발생했을 때, error 발생 이후의 code는 실행되지 않음
- code가 예상치 못하게 중단되는 걸 막기 위해 try statement를 사용


- raw data에서 예상치 못한 값이 있을 때 error가 일어나는 편
    - missing data, numbers and str mixed, negative price, $\cdots$
    

### syntax

```python
*Error?
```

```python
try:
    statement1
except ErrorName: # *Error? 
    statement2
else:
    statement3 # run if exception does not occur (optional)
    
finally: 
    statement4 # run whether exception occurrs or not (optional)
```

```python
for x in range(5):
    try:
        print(int(x))
    except ValueError:
        pass
    else:
        print('no exception')
```

```python
mylist = [1,2,3, 'a', 5]
for x in mylist:
#     print(x)
    print(int(x))
```

```python
mylist = [1,2,3,'a',5]
for x in mylist:
    try:
        print(int(x))
    except ValueError: # 
        print(f'{x} is problematic.')
#     finally:
#         print('run this part regardless of whether an exception occurs or not.')
```

## except statement with wrong error type

```python
mylist = [1,2,3,'a',5]
for x in mylist:
    try:
        print(int(x))
    except IndexError:
        print(f'{x} is problematic.')
```

## without specifying error type
- error type을 명시하는 게 권장됨
- 명시하지 않아도 코드는 실행되나, 내가 예측하지 못한 곳에서 error가 발생하는 것을 알아차릴 수 없음

```python
mylist = [1,2,3,'a',5]
for x in mylist:
    try:
        print(int(x))
    except: # not recommended (not specifying error type)
        pass
```

```python
mylist = [1,2,3,'a',5]
for x in mylist:
    try:
        print(int(x))
    except Exception as ex:
        print('error:', ex)
```
