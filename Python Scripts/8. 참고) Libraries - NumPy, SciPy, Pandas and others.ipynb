### flatten vs ravel
- flatten은 copy를, ravel은 copy를 하지 않음
    - 원래의 object를 변형시, ravel은 같이 변함
    - flatten은 변하지 않음

```python
import numpy as np
a = np.array([[1,2],[3,4]])
a1 = a.flatten()
a2 = a.ravel()
print(f'Before change. flatten: {a1}, ravel: {a2}')
a[0][0] = 20
print(f'After 1st change. flatten: {a1}, ravel: {a2}')
a[1][1] = 40
print(f'After 2nd change. flatten: {a1}, ravel: {a2}')
```

### shape manipulation

```python
x1 = np.array([[0,1,2], [2,3,4]])
print(x1)
print('-'*80)

print(x1.shape)
print(x1.size) # number of elements in array
print(len(x1))
print(len(x1.T)) # close to np.transpose(x1)

print('-'*80)
print(f'x1.reshape([3,2]):\n {x1.reshape([3,2])}, \n\n x1 after reshape:\n {x1}')
print('-'*80)
print(f'x1.resize([3,2]):\n {x1.resize([3,2])}, \n\n x1 after reshape:\n {x1}') # resize returns None
```

### np.linspace()
- returns evenly spaced numbers over a specified interval

```python
print(np.linspace(0, 10, 9))
print(np.linspace(start=0, stop=10, num=9))
print(np.linspace(stop=10, start=0, num=11)) # increasing
print('-'*80)
print(np.linspace(10, 0, 11)) # decreasing
```

### axis of NumPy
- axis=0은 row-wise (or vertical)
- axis=1은 column-wise (or horizontal)
- axis=0은 flatten

```python
# np.concatenate() joins a sequence of arrays along an existing axis
# default: axis=0
x1 = np.identity(2)
x2 = np.array([[1,2], [2,1]])
print('axis=0:\n', np.concatenate((x1, x2), axis=0), '\n') # row-wise concatenation (vertical stack)
print('axis=1:\n', np.concatenate((x1, x2), axis=1), '\n') # column-wise concatenation (horizontal stack)
print('axis=None:\n', np.concatenate((x1, x2), axis=None), '\n') # flatten
```

### np.eye()
- np.identity()는 identity matrix만 제공
- np.eye()는 off-diagonal에 1을 저장하는 등의 추가적인 기능 제공

```python
# identity matrix
print(np.eye(3), '\n') # equivalent to np.identity(3)
print(np.eye(3, 2), '\n')
print(np.eye(2, 3), '\n')

print(np.eye(3, k=1), '\n') # upper off-diagonal
print(np.eye(3, k=-1), '\n') # lower off-diagonal
```

### np.multiply(), np.dot(), np.matmul()
- x\*y는 element-wise 곱을 반환
- np.multiply()는 matrix의 element-wise 곱을 반환
- np.matmul()은 matrix product를 반환
- np.dot()는 dot product를 반환
    - 2차원 배열인 행렬 a, b에서는 matmul()과 결과가 같음
    - 그 외에는 결과가 다를 수 있음

```python
import numpy as np
x = np.identity(2) # equivalent to np.eye(2)
y = np.array([[2, 3], [1, 2]])
print(type(x))
print(type(y))

print(f'x:{x}\n')
print(f'y:{y}\n')

print('-'*80)
print(f'x*y:\n {x*y}\n') # element-wise operation
print(f'np.multiply():\n {np.multiply(x,y)}\n') # element-wise operation

print('-'*80)
print(f'np.dot(x,y):\n {np.dot(x,y)}\n')
print(f'x*y:\n {x.dot(y)}\n')

print('-'*80)
print(f'np.matmul():\n x*y:{np.matmul(x,y)}\n')
```

### np.repeat() and np.tile()
- np.repeat(a, repeats)은 array a의 each element를 repeats회만큼 반복
    - np.repeat([0, 1], 2) = [0, 0, 1, 1]


- np.tile(A, reps)은 array A를 reps회만큼 반복
    - np.tile([0, 1], 2) = [0, 1, 0, 1]

```python
x = np.array(range(3))
print(f'x: {x}\n')
print('-'*80)

# element-wise 
print(f'x.repeat(3, axis=0):\n {x.repeat(3, axis=0)}\n') # default axis
print(f'x.repeat(3, axis=1):\n {x.repeat(3, axis=0)}\n') # default axis

print('-'*80)
x = np.array([[1,2], [3, 1]])
print(f'x: {x}\n')

print(f'x.repeat(2, axis=0):\n {x.repeat(2, axis=0)}\n')
print(f'x.repeat(2, axis=1):\n {x.repeat(2, axis=1)}\n')
print('-'*80)
# array-wise
print(f'np.tile(x, (1,2)):\n {np.tile(x, (1,2))}\n')
print(f'np.tile(x, (2,1)):\n {np.tile(x, (2,1))}\n')
```

## Optimization

```python
# ex) 
import matplotlib.pyplot as plt

def myQuad(x):
    return (x-1)**2

x = np.linspace(start=-10, stop=10, num=50)
plt.plot(x, myQuad(x))
plt.grid() # grid를 print
plt.xlabel('x')
plt.ylabel('f(x)=(x-1)**2')
plt.show()
```

## finding solution with scipy.optimize
- myQuad 함수의 해를 scipy.optimize.minimize()를 통해 구할 수 있음
- f = 0의 형태로 만든 뒤, scipy.optimize.minimize(f, x0=init)을 사용해 찾음
- init 값에 따라 해찾기에 걸리는 시간이 달라짐
- 함수가 복잡할수록, init값이 실제 해에서 멀수록 해를 구하기 힘들어짐

```python
from scipy.optimize import minimize
x_ = minimize(myQuad, x0=-1)
print(type(x_), '\n')
print(x_)
```

- x: ndarray
    - The solution of the optimization.
- success: bool
    - Whether or not the optimizer exited successfully.
- nit: int
    - Number of iterations performed by the optimizer.

```python
from scipy.optimize import minimize
for x0 in [-5, -3, 1, 2, 3]:
    x_ = minimize(myQuad, x0=-1)
    # print(type(x_), '\n')
    print(f"x0={x0}, x={x_.x[0]}")
```

```python
def myPoly3(x):
    return (x-2)*x*(x+2)

x = np.linspace(start=-5, stop=5, num=50)
plt.plot(x, myPoly3(x))
plt.xlabel('x')
plt.ylabel('f(x)=(x-2)x(x+2)')
plt.grid()
plt.show()
```

```python
# ex) 4th-degree polynomial
def myPoly4(x):
    return (x-3)*(x-1)*(x+1)*(x+3)

x = np.linspace(start=-5, stop=5, num=50)
plt.plot(x, myPoly4(x))
plt.xlabel('x')
plt.ylabel('f(x)=(x-3)(x-1)(x+1)(x+3)')
plt.grid()
plt.show()
```

```python
for x in [-3, -1, 1, 3]:
    print(f"x={x}, myPoly4(x)={myPoly4(x)}")
```

### solution depends on initial value
- myPoly4 함수의 해는 -3, -1, 1, 3
- initial value가 아래와 같은 경우, 전부 해를 찾는 데에 실패함

```python
x0 = [-2, -1, 0., 1, 2]
for x0_ in x0:
    x_ = minimize(myPoly4, x0=x0_)
    print(f'x0: {x0_}')
    print(f'success: {x_.success}')
    print(f'number of iteration: {x_.nit}')
    print(f'solution: {x_.x[0]:.2f}\n')
```

- tol: float, optional
    - Tolerance for termination
    
- objective function의 오차
- tol값이 크면 최적화가 빨리 끝나는 대신 해가 부정확
- tol값이 크면 해가 정확해지는 대신 최적화가 오래 걸림

- 해를 찾지 못하는 이유는 최적화 알고리즘이 내부적으로 미분을 사용하기 때문
- 아래에서 기울기가 0인 지점들을 찾으면 다음과 같음

```python
import sympy as sym
x = sym.Symbol('x')
print(sym.solve((x-3)*(x-1)*(x+1)*(x+3)))

print( sym.solve(sym.diff((x-3)*(x-1)*(x+1)*(x+3)), x) )
```

- 주어진 x에서 미분값을 계산한 뒤, 그 미분값의 크기만큼 x 방향으로 진행
    - 개념적으로 $x_{n+1} = x_{n} - f'(x_{n})$ 에 해당
- 기울기가 크면 $x_{n}$의 변동이 심함 $\Rightarrow$ 해를 찾기가 어려움
- 주어진 x에서 미분 값이 0이면 더 이상 iteration을 진행하지 않음

```python
x0 = [-np.sqrt(5), -1, 0, 1, 2]
for x0_ in x0:
    x_ = minimize(myPoly4, x0=x0_, method='BFGS', tol=1e-8)
    print(f'x0: {x0_}')
    print(f'success: {x_.success}')
    print(f'number of iteration: {x_.nit}')
    print(f'solution: {x_.x[0]:.2f}\n')
```

# SymPy
- Symbolic math
- 미분, 적분, 해 찾기 등을 x, y등의 변수를 이용해 사용가능

## sympy.subs()
- subs()를 통해 한 expression을 다른 expression으로 바꿀 수 있음
    - x를 y로 변환하거나, x에 numeric value를 substitute 가능

```python
import sympy as sym
x = sym.symbols('x')
y = sym.symbols('y')
expr = sym.cos(x) + 1
print(expr.subs(x, y)) # does not overwrite to expr
print(f'expr after subs: {expr}')
print(expr.subs(x, 0))
```

## sympy.evalf()
- sympy.evalf()를 통해 eval()처럼 expression을 evaluate 가능

```python
import sympy as sym
a = sym.Rational(1, 2)
print(type(a), a)

print(sym.pi * 2)
print(sym.pi.evalf())

print('-'*80)
print((sym.pi + sym.exp(1))) # not evaluated yet
print((sym.pi + sym.exp(1)).evalf()) # evaluated

print('-'*80)
print(sym.oo) # infinity
print(-sym.oo) # -infinity
```

### sympy.diff(), .integrate(), sympy.solve(), sympy. solveset()
- sympy.diff()를 통해 expression을 미분 가능
- sympy.integrate()을 통해 expression을 적분 가능
- sympy.solve(), solveset()을 통해 expression의 해를 찾는 것이 가능
    - 대부분의 경우 함수의 꼴이 복잡하고, 해의 존재성 혹은 유일성이 알려져 있지 않음
    - 일반적인 경우는 numerical method (e.g. scipy.optimize.minimize)를 통해 해를 구함
    
    
- solve vs solveset
    - solve()에 문제가 있어 solveset()을 새로 개발하는 중 (https://docs.sympy.org/latest/modules/solvers/solveset.html#what-s-wrong-with-solve)
    - solveset()이 할 수 없지만 solve()가 할 수 있는 기능들이 있음

```python
x = sym.Symbol('x')
y = sym.Symbol('y')

print(x + y + x - y + y)

print('-'*80)

print(sym.diff(sym.sin(x), x))
eq = (x + y)**2
print(eq)
print(sym.diff(eq, x))

print('-'*80)
print(sym.integrate(3*x**2+5, x))

print('-'*80)
print(sym.solveset(x**2 - 2*x + 1, x))
print(sym.solveset((x-3)*(x-1)*(x+1)*(x+3), x))

print('-'*80)
sol = sym.solve((x - 2*y + 5, -2*x + y +3), (x, y))
print(sol) # dict
```

## MultiIndex
- document: https://pandas.pydata.org/pandas-docs/stable/user_guide/advanced.html

- index (row label)은 하나의 변수가 아닌 여러 변수의 pair로도 지정 가능
- index (or MultiIndex)를 사용하면 join/sort 등을 할 때에 보다 손쉽게 계산 가능
- DataFrame의 size가 큰 경우, index를 사용하는 것이 그렇지 않은 경우보다 더 빠름

### example

```python
import itertools
firms = ['a', 'b', 'c']
dates = pd.to_datetime(['1996-01-01', '1996-02-01', '1996-03-01'])

tup = itertools.product(firms, dates) # Cartesian product
multidx = pd.MultiIndex.from_tuples(tup, names=['firms', 'dates'])

# multidx = pd.MultiIndex.from_product([firms, dates], names=['firms', 'dates']) # same as above

print(f'multidx:\n {multidx}')

prc = np.linspace(0, len(multidx), len(multidx))

df = pd.DataFrame(data=prc, columns=['prc'], index=multidx)
display(df)
```

### Sorting a MultiIndex
- 위 예제는 firms, dates 순서로 sort됨
- MultiIndex 내의 특정 index에 대해 sort할 때 sort_index(level)를 사용 가능
    - level은 position과 keyword 모두 가능
    
- 아래의 예제를 통해 dates, firms 순서로 sort를 할 수 있음
- sort는 기존 dataframe에 덮어쓰기를 하지 않음

```python
df.sort_index(level=['dates', 'firms'])
```

### (sub) index of a MultiIndex
- get_level_values(level)를 통해 특정 level의 index 값을 get 가능
    - level에는 position과 keyword 모두 사용 가능

```python
print(df.index, '\n\n') 
print(df.index.get_level_values('firms'), '\n\n') 
# print(df.index.get_level_values(0)) # same as above

print(df.index.get_level_values('dates'), '\n\n') 
```
